# overall diagnosis trend

```python
import pandas as pd
import numpy as np
import holidays
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor

# Ensure the date column is in datetime format
data['date'] = pd.to_datetime(data['date'])

# Preprocessing function
def preprocess_data(df):
    # Aggregate data to get daily diagnosis count
    time_series = df.groupby('date').agg({'diagnosis': 'count'}).reset_index()
    time_series.rename(columns={'diagnosis': 'diagnosis_count'}, inplace=True)

    # Temporal features
    time_series['month'] = time_series['date'].dt.month
    time_series['day_of_week'] = time_series['date'].dt.dayofweek
    time_series['year'] = time_series['date'].dt.year
    time_series['is_weekend'] = time_series['day_of_week'].isin([5, 6]).astype(int)

    india_holidays = holidays.India(years=range(time_series['year'].min(), time_series['year'].max() + 1))
    time_series['is_holiday'] = time_series['date'].isin(india_holidays).astype(int)

    # Lag and rolling features
    for lag in range(1, 8):
        time_series[f'diagnosis_count_lag_{lag}'] = time_series['diagnosis_count'].shift(lag)

    time_series['rolling_mean_3'] = time_series['diagnosis_count'].rolling(window=3).mean()
    time_series['rolling_mean_7'] = time_series['diagnosis_count'].rolling(window=7).mean()

    return time_series.dropna()

# Preprocess data
time_series = preprocess_data(data)

# Train-test split (last 7 months as test data)
max_date = time_series['date'].max()
train_data = time_series[time_series['date'] <= max_date - pd.DateOffset(months=7)]
test_data = time_series[time_series['date'] > max_date - pd.DateOffset(months=7)]

# Feature selection
features = ['month', 'day_of_week', 'year', 'is_weekend', 'is_holiday'] + \
           [f'diagnosis_count_lag_{i}' for i in range(1, 8)] + \
           ['rolling_mean_3', 'rolling_mean_7']

X_train, y_train = train_data[features], train_data['diagnosis_count']
X_test, y_test = test_data[features], test_data['diagnosis_count']

# Scaling the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# XGBoost model with hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

xgb_model = XGBRegressor(objective='reg:squarederror', random_state=42)
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, scoring='r2', cv=3, n_jobs=-1)

grid_search.fit(X_train_scaled, y_train)

# Best model from GridSearchCV
best_xgb = grid_search.best_estimator_
y_train_pred = best_xgb.predict(X_train_scaled)
y_test_pred = best_xgb.predict(X_test_scaled)

# Store results
results = {
    "Train R2": r2_score(y_train, y_train_pred),
    "Test R2": r2_score(y_test, y_test_pred)
}
print("Model Performance:", results)

# Combine actual and predicted data for visualization
train_data['predicted'] = y_train_pred
test_data['predicted'] = y_test_pred
full_data = pd.concat([train_data, test_data])[['date', 'diagnosis_count', 'predicted']]

# Plot results
fig, ax = plt.subplots(figsize=(15, 6))
ax.set_xlabel("Date", fontsize=16)
ax.set_ylabel("Diagnosis Count", fontsize=16)

ax.plot(train_data['date'], train_data['diagnosis_count'], label="Training Actual", color="blue", linewidth=1.5)
ax.plot(test_data['date'], test_data['diagnosis_count'], label="Test Actual", color="green", linestyle="-", linewidth=1.5)
ax.plot(test_data['date'], test_data['predicted'], label="Test Predicted", color="red", linestyle="--", linewidth=1.5)

# Improve x-axis formatting
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)

plt.xticks(rotation=45, ha="right")
ax.legend()
ax.grid()
plt.tight_layout()
plt.savefig("forecast_diagnosis_count.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import numpy as np
import holidays
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor

# Ensure the date column is in datetime format
data['date'] = pd.to_datetime(data['date'])

# Preprocessing function for drug forecasting
def preprocess_drug_data(df):
    # Combine all drug columns into a single column
    drug_columns = [f'drug_{i}' for i in range(1, 10)]
    df['total_drugs'] = df[drug_columns].notna().sum(axis=1)

    # Aggregate drug count per day
    time_series = df.groupby('date').agg({'total_drugs': 'sum'}).reset_index()

    # Temporal features
    time_series['month'] = time_series['date'].dt.month
    time_series['day_of_week'] = time_series['date'].dt.dayofweek
    time_series['year'] = time_series['date'].dt.year
    time_series['is_weekend'] = time_series['day_of_week'].isin([5, 6]).astype(int)

    india_holidays = holidays.India(years=range(time_series['year'].min(), time_series['year'].max() + 1))
    time_series['is_holiday'] = time_series['date'].isin(india_holidays).astype(int)

    # Lag and rolling features
    for lag in range(1, 8):
        time_series[f'total_drugs_lag_{lag}'] = time_series['total_drugs'].shift(lag)

    time_series['rolling_mean_3'] = time_series['total_drugs'].rolling(window=3).mean()
    time_series['rolling_mean_7'] = time_series['total_drugs'].rolling(window=7).mean()

    return time_series.dropna()

# Preprocess data
drug_time_series = preprocess_drug_data(data)

# Train-test split (last 7 months as test data)
max_date = drug_time_series['date'].max()
train_data = drug_time_series[drug_time_series['date'] <= max_date - pd.DateOffset(months=7)]
test_data = drug_time_series[drug_time_series['date'] > max_date - pd.DateOffset(months=7)]

# Feature selection
features = ['month', 'day_of_week', 'year', 'is_weekend', 'is_holiday'] + \
           [f'total_drugs_lag_{i}' for i in range(1, 8)] + \
           ['rolling_mean_3', 'rolling_mean_7']

X_train, y_train = train_data[features], train_data['total_drugs']
X_test, y_test = test_data[features], test_data['total_drugs']

# Scaling the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# XGBoost model with hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

xgb_model = XGBRegressor(objective='reg:squarederror', random_state=42)
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, scoring='r2', cv=3, n_jobs=-1)

grid_search.fit(X_train_scaled, y_train)

# Best model from GridSearchCV
best_xgb = grid_search.best_estimator_
y_train_pred = best_xgb.predict(X_train_scaled)
y_test_pred = best_xgb.predict(X_test_scaled)

# Store results
results = {
    "Train R2": r2_score(y_train, y_train_pred),
    "Test R2": r2_score(y_test, y_test_pred)
}
print("Model Performance:", results)

# Combine actual and predicted data for visualization
train_data['predicted'] = y_train_pred
test_data['predicted'] = y_test_pred
full_data = pd.concat([train_data, test_data])[['date', 'total_drugs', 'predicted']]

# Plot results
fig, ax = plt.subplots(figsize=(15, 6))
ax.set_xlabel("Date", fontsize=16)
ax.set_ylabel("Total Drug Count", fontsize=16)

ax.plot(train_data['date'], train_data['total_drugs'], label="Training Actual", color="blue", linewidth=1.5)
ax.plot(test_data['date'], test_data['total_drugs'], label="Test Actual", color="green", linestyle="-", linewidth=1.5)
ax.plot(test_data['date'], test_data['predicted'], label="Test Predicted", color="red", linestyle="--", linewidth=1.5)

# Improve x-axis formatting
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)

plt.xticks(rotation=45, ha="right")
ax.legend()
ax.grid()
plt.tight_layout()
plt.savefig("forecast_drug_count.png", dpi=300)
plt.show()
```
