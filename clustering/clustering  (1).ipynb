# K-means clustering using drug, diagnois , age group and city

```python
import pandas as pd
import ast
from gensim.models import Word2Vec
from sklearn.preprocessing import OneHotEncoder
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns


# Parse the 'diagnosis' column if it's stored as a string representation of a list
data['diagnosis'] = data['diagnosis'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Fill missing values in the drug columns with 'unknown'
drug_columns = [col for col in data.columns if col.startswith('drug_')]
data[drug_columns] = data[drug_columns].fillna('unknown')

# Remove rows with missing diagnoses
data = data.dropna(subset=['diagnosis'])

# Tokenize the diagnosis
data['tokenized_diagnosis'] = data['diagnosis'].apply(lambda x: x if isinstance(x, list) else str(x).split())

# Step 1: Create Word2Vec embeddings for 'diagnosis'
tokenized_diagnoses = data['tokenized_diagnosis'].tolist()

# Train the Word2Vec model for diagnoses
w2v_model = Word2Vec(sentences=tokenized_diagnoses, vector_size=100, window=5, min_count=1, workers=4)

# Generate embeddings for each diagnosis
def get_embedding(words):
    return w2v_model.wv[words].mean(axis=0) if words else [0] * w2v_model.vector_size

data['diagnosis_embedding'] = data['tokenized_diagnosis'].apply(get_embedding)

# Step 2: Combine all drug columns into a list for each row
data['drug_sequence'] = data[drug_columns].apply(
    lambda row: [drug for drug in row if isinstance(drug, str) and pd.notna(drug) and drug != 'unknown'],
    axis=1
)

# Step 3: Train Word2Vec on tokenized drugs
drug_sequences = data['drug_sequence'].tolist()
drug_w2v_model = Word2Vec(sentences=drug_sequences, vector_size=50, window=3, min_count=1, workers=4)

# Generate embeddings for each row
def get_drug_embedding(drugs):
    return drug_w2v_model.wv[drugs].mean(axis=0) if drugs else [0] * drug_w2v_model.vector_size

data['drug_embedding'] = data['drug_sequence'].apply(get_drug_embedding)

# Combine diagnosis and drug embeddings
diagnosis_embeddings = np.vstack(data['diagnosis_embedding'].values)
drug_embeddings = np.vstack(data['drug_embedding'].values)
X = np.hstack((diagnosis_embeddings, drug_embeddings))

# Step 4: One-Hot Encoding for categorical variables (Age group, Gender, Region)
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_features = encoder.fit_transform(data[['age_group', 'city']])  # Age, Region
X = np.hstack((X, encoded_features))
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Elbow Method for optimal number of clusters
inertia = []
k_range = range(1, 11)  # Try k = 1 to 10 clusters

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)

# Save the plot as an image
filename = "elbow_plot.png"
plt.savefig(filename, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()



```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Step 5: Fit KMeans with the optimal number of clusters
optimal_k = 3  # Change this based on the elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['cluster'] = kmeans.fit_predict(X)

# Step 7: Reduce dimensions for visualization (PCA)
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Add reduced dimensions to the dataset
data['pca_x'] = X_reduced[:, 0]
data['pca_y'] = X_reduced[:, 1]

# Plot clusters using PCA visualization
plt.figure(figsize=(10, 7))
sns.scatterplot(data=data, x='pca_x', y='pca_y', hue='cluster', palette='viridis', s=50)

# Save the plot as an image
filename = "cluster_visualization.png"
plt.savefig(filename, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from collections import Counter

# Function to format large numbers
def format_large_numbers(value, _):
    if value >= 1_000_000:
        return f'{value / 1_000_000:.0f}M'
    elif value >= 1_000:
        return f'{value / 1_000:.0f}K'
    return str(int(value))

# Step 1: Clean the dataset
data['flat_diagnosis'] = data['diagnosis'].apply(lambda x: ' | '.join([d for d in x if d.lower() not in ['unknown', 'specify in advice']]) if isinstance(x, list) else str(x))

# Step 2: Get top 5 drugs for each cluster (excluding 'unknown')
top_drugs_data = []
for cluster in data['cluster'].unique():
    cluster_drugs = data[data['cluster'] == cluster]['drug_sequence'].explode()
    cluster_drugs = cluster_drugs[cluster_drugs.str.lower() != 'unknown']  # Remove 'unknown' from drugs
    drug_counts = Counter(cluster_drugs).most_common(5)

    for drug, count in drug_counts:
        top_drugs_data.append({'Cluster': f'Cluster {cluster}', 'Drug': drug, 'Count': count})

top_drugs_df = pd.DataFrame(top_drugs_data)

# Step 3: Plotting
fig, axes = plt.subplots(2, 2, figsize=(40, 20))
sns.set(style='whitegrid')

# Font sizes
label_font = 30
tick_font = 26
legend_font = 26
annotation_font = 34

# Labels for the bottom of each plot
plot_labels = [
    "(a) Age Group per Cluster",
    "(b) Top 5 Diagnoses per Cluster",
    "(c) Top 5 Drugs per Cluster",
    "(d) City Distribution per Cluster"
]

# Age group distribution
age_group_counts = data.groupby(['cluster', 'age_group']).size().reset_index(name='count')
sns.barplot(data=age_group_counts, x='age_group', y='count', hue='cluster', ax=axes[0, 0])
axes[0, 0].set_ylabel('Count', fontsize=label_font)
axes[0, 0].set_xlabel('Age Group', fontsize=label_font)
axes[0, 0].legend(title='Cluster', fontsize=legend_font, title_fontsize=legend_font)
axes[0, 0].tick_params(axis='x', rotation=45, labelsize=tick_font)
axes[0, 0].tick_params(axis='y', labelsize=tick_font)
axes[0, 0].yaxis.set_major_formatter(ticker.FuncFormatter(format_large_numbers))

# Top diagnoses distribution (excluding 'unknown' & 'specify in advice')
diagnosis_counts = data.groupby(['cluster', 'flat_diagnosis']).size().reset_index(name='count')
diagnosis_counts = diagnosis_counts[~diagnosis_counts['flat_diagnosis'].isin(['unknown', 'specify in advice'])]
top_diagnoses = diagnosis_counts.groupby('cluster').apply(lambda x: x.nlargest(5, 'count')).reset_index(drop=True)

sns.barplot(data=top_diagnoses, x='cluster', y='count', hue='flat_diagnosis', dodge=True, ax=axes[0, 1])
axes[0, 1].set_ylabel('Count', fontsize=label_font)
axes[0, 1].set_xlabel('Cluster', fontsize=label_font)
axes[0, 1].legend(title='Diagnosis', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=legend_font, title_fontsize=legend_font)
axes[0, 1].tick_params(axis='x', labelsize=tick_font)
axes[0, 1].tick_params(axis='y', labelsize=tick_font)
axes[0, 1].yaxis.set_major_formatter(ticker.FuncFormatter(format_large_numbers))

# Top drugs distribution
sns.barplot(data=top_drugs_df, x='Cluster', y='Count', hue='Drug', ax=axes[1, 0], palette='viridis')
axes[1, 0].set_ylabel('Count', fontsize=label_font)
axes[1, 0].set_xlabel('Cluster', fontsize=label_font)
axes[1, 0].legend(title='Drug', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=legend_font, title_fontsize=legend_font)
axes[1, 0].tick_params(axis='x', labelsize=tick_font)
axes[1, 0].tick_params(axis='y', labelsize=tick_font)
axes[1, 0].yaxis.set_major_formatter(ticker.FuncFormatter(format_large_numbers))

# City distribution
city_counts = data.groupby(['cluster', 'city']).size().reset_index(name='count')
sns.barplot(data=city_counts, x='cluster', y='count', hue='city', ax=axes[1, 1])
axes[1, 1].set_ylabel('Count', fontsize=label_font)
axes[1, 1].set_xlabel('Cluster', fontsize=label_font)
axes[1, 1].legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=legend_font, title_fontsize=legend_font)
axes[1, 1].tick_params(axis='x', labelsize=tick_font)
axes[1, 1].tick_params(axis='y', labelsize=tick_font)
axes[1, 1].yaxis.set_major_formatter(ticker.FuncFormatter(format_large_numbers))

# Add labels below each plot
for ax, label in zip(axes.flat, plot_labels):
    ax.text(0.5, -0.3, label, fontsize=annotation_font,
            ha='center', va='center', transform=ax.transAxes)

# Adjust layout
plt.tight_layout()

# Save the plot
output_path = 'analysis_plot_filtered.png'
plt.savefig(output_path, bbox_inches='tight', dpi=300)

# Show the plot
plt.show()
```

# Kmeans using drug,diagnosis , age , city,date and appointment time

```python
import ast
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (Assuming `data` is your DataFrame)
# data = pd.read_csv("your_file.csv")  # Uncomment if loading from CSV

# Step 1: Handle Missing Diagnoses
data['diagnosis'] = data['diagnosis'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
data = data.dropna(subset=['diagnosis'])  # Remove rows where diagnosis is missing
data['tokenized_diagnosis'] = data['diagnosis'].apply(lambda x: x if isinstance(x, list) else str(x).split())

# Step 2: Train Word2Vec for Diagnosis
tokenized_diagnoses = data['tokenized_diagnosis'].tolist()
w2v_model = Word2Vec(sentences=tokenized_diagnoses, vector_size=100, window=5, min_count=1, workers=4)

# Function to get diagnosis embeddings
def get_embedding(words):
    if not words:
        return np.zeros(w2v_model.vector_size)  # Avoid NaN
    return w2v_model.wv[words].mean(axis=0)

data['diagnosis_embedding'] = data['tokenized_diagnosis'].apply(get_embedding)

# Step 3: Process Drug Data
drug_columns = [col for col in data.columns if col.startswith('drug_')]
data[drug_columns] = data[drug_columns].fillna('unknown')  # Fill missing drugs with 'unknown'
data['drug_sequence'] = data[drug_columns].apply(lambda row: [drug for drug in row if drug != 'unknown'], axis=1)

# Step 4: Train Word2Vec for Drug Sequences
drug_sequences = data['drug_sequence'].tolist()
drug_w2v_model = Word2Vec(sentences=drug_sequences, vector_size=50, window=3, min_count=1, workers=4)

# Function to get drug embeddings
def get_drug_embedding(drugs):
    if not drugs:
        return np.zeros(drug_w2v_model.vector_size)  # Avoid NaN
    return drug_w2v_model.wv[drugs].mean(axis=0)

data['drug_embedding'] = data['drug_sequence'].apply(get_drug_embedding)

# Step 5: Encoding City (Already Encoded as `city_encoded` in your data)
# No change needed

# Step 6: Encoding Date & Appointment Time
data['date'] = pd.to_datetime(data['date'])
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day

# Convert appointment time into minutes since midnight
data['appointmenttime'] = pd.to_datetime(data['appointmenttime'], format='%H:%M:%S', errors='coerce')
data['appointment_minutes'] = data['appointmenttime'].dt.hour * 60 + data['appointmenttime'].dt.minute

# Handle NaN in time column
data['appointment_minutes'] = data['appointment_minutes'].fillna(-1)  # -1 for missing values

# Step 7: One-Hot Encoding Categorical Data
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
encoded_features = encoder.fit_transform(data[['district']])  # Encoding 'district' (excluding city, already encoded)

# Step 8: Combine All Features for Clustering
diagnosis_embeddings = np.vstack(data['diagnosis_embedding'].values)
drug_embeddings = np.vstack(data['drug_embedding'].values)

X = np.hstack((
    diagnosis_embeddings,
    drug_embeddings,
    data[['age', 'year', 'month', 'day', 'appointment_minutes', 'city_encoded']].fillna(-1).values,  # -1 for NaNs
    encoded_features
))

# Step 9: Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 10: Determine Optimal Clusters Using Elbow Method
wcss = []
for i in range(2, 11):  # Checking clusters from 2 to 10
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot Elbow Method Graph
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal Clusters')
plt.show()

# Step 11: Apply KMeans Clustering with Optimal K
optimal_k = 3  # Choose based on the elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
data['cluster'] = kmeans.fit_predict(X_scaled)

# Step 12: Visualizing Clusters (PCA for Dimensionality Reduction)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
data['pca_x'] = X_pca[:, 0]
data['pca_y'] = X_pca[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['pca_x'], y=data['pca_y'], hue=data['cluster'], palette='viridis')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('KMeans Clustering Visualization')
plt.legend(title='Cluster')
plt.show()
```
